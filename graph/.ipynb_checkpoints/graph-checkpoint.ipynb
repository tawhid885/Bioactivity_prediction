# MD simulation trajectory analysis

```python
import matplotlib.pyplot as plt
import numpy as np
```

# RMSD 

```python
def load_rmsd(file_name):
    time = []
    rmsd = []
    with open(file_name) as file:
        for line in file:
            if not line.startswith(('@', '#')):
                t, r = map(float, line.split())
                time.append(t)
                rmsd.append(r)
    return time, rmsd
```

```python
# Load RMSD data
time_protein, protein_rmsd = load_rmsd('protein_rmsd.xvg')
time_complex, complex_rmsd = load_rmsd('complex_rmsd.xvg')
```

```python
protein_average_rmsd = sum(protein_rmsd)/len(protein_rmsd)
complex_average_rmsd = sum(complex_rmsd)/len(complex_rmsd)
```

```python
protein_average_rmsd
```

```python
complex_average_rmsd
```

## RMSD 

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time_complex, complex_rmsd, label='Protein-Ligand Complex', color='red')
plt.plot(time_protein, protein_rmsd, label='Protein_rmsd', color='blue')


# plt.title('RMSD Comparison: Protein vs Protein-Ligand Complex')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.legend()
plt.savefig('rmsd_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

## RMSF

```python
# Function to read .xvg files
def read_xvg(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            if not line.startswith(('#', '@')):  # Ignore header lines
                columns = line.split()
                data.append([float(columns[0]), float(columns[1])])
    return np.array(data)
```

```python
# Load data from the RMSF .xvg files
protein_rmsf = read_xvg('protein_rmsf.xvg')
complex_rmsf = read_xvg('complex_rmsf.xvg')
```

```python
# Extract residue indices and RMSF values
protein_residues = protein_rmsf[:, 0]
protein_rmsf_values = protein_rmsf[:, 1]
complex_residues = complex_rmsf[:, 0]
complex_rmsf_values = complex_rmsf[:, 1]
```

```python
# Calculate average RMSF values
protein_avg_rmsf = np.mean(protein_rmsf_values)
complex_avg_rmsf = np.mean(complex_rmsf_values)
```

```python
protein_avg_rmsf
```

```python
complex_avg_rmsf
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(complex_residues, complex_rmsf_values, label='Protein-Ligand Complex', color='red')
plt.plot(protein_residues, protein_rmsf_values, label='Protein', color='blue')


# plt.title('RMSF')
plt.xlabel('Residue Index')
plt.ylabel('RMSF (nm)')
plt.legend()
plt.savefig('rmsf_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

## HYDROGEN BOND

```python
def load_xvg(filename):
    time = []
    hbonds = []
    
    with open(filename, 'r') as file:
        for line in file:
            if not line.startswith(('@', '#')):
                cols = line.split()
                time.append(float(cols[0]))  # Time
                hbonds.append(float(cols[1]))  # Number of hydrogen bonds
    
    return time, hbonds
```

```python
# Load the data
time, hbonds = load_xvg('hb.xvg')
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, hbonds, color='blue', linestyle='-', marker=None)
plt.xlabel('Time (ns)')
plt.ylabel('Number of Hydrogen Bonds')
# plt.title('Hydrogen Bonds Over Time')
plt.grid(False)
plt.savefig('h_bond.png', dpi=300, bbox_inches='tight')
plt.show()
```

## SASA

```python
# Function to load .xvg data, skipping header lines
def load_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if not line.startswith(('@', '#')):
                data.append([float(x) for x in line.split()])
    return np.array(data)

# Load the sasa.xvg data
data = load_xvg('complex_sasa.xvg')

# Extract time (first column) and SASA values (second column)
time = data[:, 0]
sasa = data[:, 1]
```

```python
protein_data = load_xvg('complex_sasa.xvg')
complex_data = load_xvg('protein_sasa.xvg')
```

```python
protein_time = protein_data[:, 0]
protein_sasa = protein_data[:, 1]
complex_time = complex_data[:, 0]
complex_sasa = complex_data[:, 1]
```

```python
protein_average_sasa = sum(protein_sasa)/len(protein_sasa)
complex_average_sasa = sum(complex_sasa)/len(complex_sasa)
```

```python
protein_average_sasa
```

```python
complex_average_sasa
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(protein_time, protein_sasa, label='Protein', color='blue')
plt.plot(complex_time, complex_sasa, label='Protein-Ligand Complex', color='red')



# plt.title('SASA')
plt.xlabel('Time (ns)')
plt.ylabel(r'Area (nm$^2$)')
plt.legend()
plt.savefig('sasa_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Radius of gyration

```python
# Function to read .xvg files and extract time and Rg data
def read_xvg(file_path):
    time = []
    rg = []
    with open(file_path, 'r') as file:
        for line in file:
            if not line.startswith(('#', '@')):  # Ignore header lines starting with # or @
                columns = line.split()
                time.append(float(columns[0]))
                rg.append(float(columns[1]))
    return np.array(time), np.array(rg)
```

```python
# Load data from the .xvg files
protein_time, protein_rg = read_xvg('protein_gyrate.xvg')
complex_time, complex_rg = read_xvg('complex_gyrate.xvg')
```

```python
protein_average_gyrate = sum(protein_rg)/len(protein_rg)
complex_average_gyrate = sum(complex_rg)/len(protein_rg)
```

```python
protein_average_gyrate
```

```python
complex_average_gyrate
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(complex_time, complex_rg, label='Protein-Ligand Complex', color='red')
plt.plot(protein_time, protein_rg, label='Protein', color='blue')


# plt.title('RG')
plt.xlabel('Time (ns)')
plt.ylabel('RG (nm)')
plt.legend()
plt.savefig('gyration_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
